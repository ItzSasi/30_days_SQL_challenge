```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.window import Window

#Sample data
data = [
    (1, "2024-01-01", "North", "Laptop",   500),
    (2, "2024-01-02", "North", "Mouse",    300),
    (3, "2024-01-03", "North", "Keyboard", 700),
    (4, "2024-01-04", "North", "Monitor",  1200),
    (5, "2024-01-01", "South", "Phone",    400),
    (6, "2024-01-02", "South", "Tablet",   600),
    (7, "2024-01-03", "South", "Charger",  150),
    (8, "2024-01-04", "South", "Cable",    80),
]
cols = ["sale_id", "sale_date", "region", "product", "revenue"]
df = spark.createDataFrame(data, cols)
```

```python
window_spec = (
    Window
    .partitionBy("region")
    .orderBy("sale_date")
    .rowsBetween(Window.unboundedPreceding, Window.currentRow)
)
```

```python
result = df.withColumn(
    "running_total",
    F.sum("revenue").over(window_spec)
)
```

```python
result.orderBy("region", "sale_date").show()
```

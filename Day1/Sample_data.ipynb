```python
%sql
DROP TABLE IF EXISTS daily_sales;
CREATE TABLE daily_sales (
  sale_id   INT,
  sale_date DATE,
  region    VARCHAR(20),
  product   VARCHAR(50),
  revenue   INT
);
```

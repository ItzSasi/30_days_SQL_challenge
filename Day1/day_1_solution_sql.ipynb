### You have a daily sales table. For each row, calculate the running total of revenue partitioned by region and ordered by sale date — without using any GROUP BY aggregation.

```python
%sql
SELECT sale_date, region, product, revenue,
SUM(revenue) OVER (PARTITION BY region ORDER BY sale_date ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS running_total
FROM  daily_sales
ORDER BY 2, 1;
```
